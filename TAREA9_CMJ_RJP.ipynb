Realice la prueba de chi-cuadrado para validar la independencia del punto del ciclo de negocios en que se encuentre la econom√≠a.

```python
import pandas as pd
data = pd.DataFrame(index=['alta','mediana','baja','total_'],columns=['pico','abajo','subiendo','bajando','total'])
data.iloc[0,:-1] = [20,30,20,30]
data.iloc[1,:-1] = [7,40,8,5]
data.iloc[2,:-1] = [3,30,2,5]
data['total'] = data.sum(axis=1)
data.loc['total_',:] = data.sum(axis=0)
data
```

```python
import numpy as np
P = data.iloc[0,4]/data.iloc[-1,-1]
f_o = data.loc[data.index!='total_',data.columns!='total'].values.flatten()
f_e = np.concatenate([data.iloc[-1,:-1]*P,data.iloc[-1,:-1]*(1-P)])

P
# data.index
```

```python
import numpy as np
import scipy.stats as st     # Librer√≠a estad√≠stica
import matplotlib.pyplot as plt

#### M√©todo chi cuadrado
F_obse = f_o #[20,30,20,30,7,40,8,5,3,30,2,5]
F_espe = f_e #[15,50,15,20,15,50,15,20]

F_espe
```

```python
x2 = st.chisquare(F_obse,F_espe,ddof=4)
print('Valor de chi cuadrado = ',list(x2)[0],',p-value de la prueba=',list(x2)[1])

Ji = st.chi2.ppf(q = 0.9,df=4)
print('Estad√≠stico de Ji = ',Ji)
# st.t.interval
```

```python
x1 = st.chi2.cdf(list(x2)[0],df=3)
x1
```
